```python
last_year = date.today().year - 1
```

### Getting the close time for issues 
Parsing the information from json to get the close time for issues<br>
**time_days** - dict, structured as 
- time_days : {repo : [ ]list of issue close time(in days) for that repo}

**time_hours** - dict, structured as 
- time_hours : {repo : [ ]list of issue close time(in hours) for that repo}

```python
time_days = defaultdict(list)
time_hours = defaultdict(list)
for candidate in repos:
    candidate = candidate.lower()
    with open(candidate + '.json') as f:
        for line in f:
            line  = json.loads(line)
            ## If the category is issue and it is not a pull request
            if(line['data']['state'] == 'closed' and "pull_request" not in line["data"]):                 
                created = datetime.strptime(line['data']['created_at'],"%Y-%m-%dT%H:%M:%SZ")
                #if the issue was created last year
                if(created.year == last_year):
                    closed = datetime.strptime(line['data']['closed_at'],"%Y-%m-%dT%H:%M:%SZ")
                    
                    time =  closed - created
                    time_days[candidate].append(time.days)    #time in days
                
                    duration_in_s = time.total_seconds()
                    hours = int(divmod(duration_in_s, 3600)[0])
                    time_hours[candidate].append(hours)       #time in hours
```

### Visualizing the issue close_time for analysis for individual repos
We can visualize the distribution of closing time<br> 
- in a detailed manner - **density plots** (using time to close in hours)
- in a less detailed (just to get an upper level idea) way - **histogram plots**<br>

Also, in **histogram plots**, only issues that survive longer then 5 days are analyzed for better visualization (which would help in analyzing the general trend)

```python
#To see the distribution of closing time in a summarized way for issues surviving longer than 5 days.
for repo in repos:
    repo_ = repo.lower()
    max_val = max(time_days[repo_])  # getting max. value to better fit the histogram plot
    fig = plt.figure()
    fig.subplots_adjust(hspace=0, wspace=0)
    
    ax = fig.add_subplot(1,2,1)
    
    ax.hist(time_days[repo_], bins = int(max_val/5), color = 'blue', edgecolor = 'black', range = [5, max_val])# using bins that are 5 days in length 
    
    ymin, ymax = plt.gca().get_ylim()
    ax.set_ylim(top = 1.05*ymax)
    ax.set_xlim(right = max_val * 1.05)
    ax.set_xlabel("Time to close (in days)")
    ax.set_ylabel("Issue count")
    ax.set_title(repo + " :Close_time distr.(for issues which survive for more than 5 days)")

    fig.add_subplot(1,2,2)
    sns.kdeplot(time_hours[repo_], shade = True, clip = [0, max(time_hours[repo_])])
    plt.tight_layout()
    plt.title(repo + " :Closes_time distr (general)")
    plt.xlabel("Time to close (in hours)")
    plt.rcParams['figure.figsize'] = [12, 4]


    plt.show()
```

### Comparing the trend among repos through Side-by-side and Stacked histograms 

```python
lists = list()
for key, value in time_days.items():    
    lists.append(value)
    
colors = ['#E69F00', '#F0E442', '#009E73']
a = max(time_days['fabrik'])
b = max(time_days['origami'])
c = max(time_days['evalai'])
max_val = max(a,b,c)
plt.hist(lists, bins = int(max_val/20), normed = True,
         color = colors, label=repos)
plt.legend()
plt.xlabel("Closing time in days")
plt.ylabel("Normalized Issue Count")
plt.title("Side-by-side histogram with three repos")
#plt.xlim(right = 1.05*max_val)
plt.tight_layout()
plt.show()

plt.hist(lists, bins = int(max_val/10), stacked=True, color = colors, label=repos)
plt.legend()
plt.xlabel("Closing time in days")
plt.ylabel("Issue Count")
plt.title("Stacked histogram for three repos")
plt.tight_layout()
plt.show()
```

### Comparing the trend via Violinplot
[Violinplot](https://seaborn.pydata.org/generated/seaborn.violinplot.html) - The width of the violin denotes the density around that value.

```python
sns.set(context="paper", style="ticks")
ax = sns.violinplot(data = lists, palette = colors)
ax.set_xticklabels(['Origami', 'Fabrik', 'EvalAI'])
plt.rcParams['figure.figsize'] = [8, 7]
plt.xlabel("Repo")
plt.ylim(bottom = 0, top = 450)
plt.ylabel("Issue Close time (in days)")
plt.show()
```

### Visualizing through boxplots
More can be found about boxplots [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.boxplot.html)

```python
a = plt.boxplot(lists, labels = ['Origami', 'Fabrik', 'EvalAI'], whis = 1, patch_artist = True, )
for patch, color in zip(a['boxes'], colors):
       patch.set_facecolor(color)
plt.xlabel("Repo")
plt.ylim(bottom = 0, top = 200)
plt.ylabel("Issue Close time (in days)")
plt.show()
```
